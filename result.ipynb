```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('result/result.csv')
```

```python


# 서브플롯 생성
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))


# 첫 번째 서브플롯: Train Accuracy
axs[0].plot(df['epoch'], df['train_acc'], label='Train Accuracy', color='blue')
axs[0].set_title('Train Accuracy')
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('Accuracy')
axs[0].legend()

# 두 번째 서브플롯: Validation Accuracy
axs[1].plot(df['epoch'], df['validation_acc'], label='Validation Accuracy', color='green')
axs[1].set_title('Validation Accuracy')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('Accuracy')
axs[1].legend()

# 세 번째 서브플롯: Train Loss
axs[2].plot(df['epoch'], df['train_loss'], label='Train Loss', color='red')
axs[2].set_title('Train Loss')
axs[2].set_xlabel('Epoch')
axs[2].set_ylabel('Loss')
axs[2].legend()

# 서브플롯 간 간격 조정
plt.suptitle("Using ResNet-101 backborn and training")
plt.tight_layout()
# 그래프 표시
plt.show()

print(f"{np.max(df['train_acc']):.2f}, {np.max(df['validation_acc']):.2f}, {np.min(df['train_loss']):.3f}")
```
