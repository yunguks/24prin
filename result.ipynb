```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train_loss_plot_all_teacher_05_17.csv')
```

```python


#  
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))


#   : Train Accuracy
axs[0].plot(df['epoch'], df['train_acc'], label='Train Accuracy', color='blue')
axs[0].set_title('Train Accuracy')
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('Accuracy')
axs[0].legend()

#   : Validation Accuracy
axs[1].plot(df['epoch'], df['validation_acc'], label='Validation Accuracy', color='green')
axs[1].set_title('Validation Accuracy')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('Accuracy')
axs[1].legend()

# #   : Train Loss
# axs[2].plot(df['epoch'], df['train_loss'], label='Train Loss', color='red')
# axs[2].set_title('Train Loss')
# axs[2].set_xlabel('Epoch')
# axs[2].set_ylabel('Loss')
# axs[2].legend()

#    
plt.suptitle("Using teacher forcing")
plt.tight_layout()
#  
plt.show()
```
